```python
import pandas as pd
```

```python
news_df = pd.read_csv("../Resources/news.csv")
news_df.drop(columns='Unnamed: 0', inplace=True)
news_df.head()
```

```python
all_fake_news_df = pd.read_csv("../Resources/Fake.csv")
all_fake_news_df.head()
```

```python
all_real_news_df = pd.read_csv("../Resources/True.csv")
all_real_news_df.head()
```

```python
all_fake_news_df.drop(columns=['subject','date'],inplace=True)
all_fake_news_df['label'] = "FAKE"
all_fake_news_df
```

```python
all_real_news_df.drop(columns=['subject','date'],inplace=True)
all_real_news_df['label'] = "REAL"
all_real_news_df
```

```python
big_news_df = pd.concat([news_df,all_fake_news_df], ignore_index=True)
```

```python
final_news_df = pd.concat([big_news_df,all_real_news_df], ignore_index=True)
final_news_df
```

```python
final_news_df.to_csv("../Resources/big_news_output.csv")
```
